```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10.5,0.5)
# figure out exponents
exp1 = np.log10(150)
exp2 = np.log10(250)
# plot it
plot(x, x**2, "ro--", x, x**exp1, "bs", x, x**exp2, "g^-")
axis([0,10,0,300])



```
