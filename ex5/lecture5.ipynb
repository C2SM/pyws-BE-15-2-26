## Important elements of Cartopy

**`class cartopy.crs.CRS(proj4_params, globe=None)`**    
Proj.4: Most powerful Cartographic Projections Library. <br />
Refer to the [Proj.4 documentation](http://trac.osgeo.org/proj/wiki/WikiStart#Documentation) to find out more about the projection parameters, e.g. [here](http://trac.osgeo.org/proj/wiki/GenParms).

**`class cartopy.crs.Globe(datum=None, ellipse='WGS84', ...)`**
The class `Globe` describes the reference system used. 

**`class cartopy.crs.Projection`**    
Defines a projected coordinate system with flat topology and Euclidean distance. This is the class you are most like going to use to make maps.

[List of Cartopy projections](http://scitools.org.uk/cartopy/docs/latest/crs/projections.html#cartopy-projections)

## Plotting on maps

Behind the scenes, the matplotlib [`Axes`](http://matplotlib.org/api/axes_api.html) object is replaced by cartopy's [GeoAxis](http://scitools.org.uk/cartopy/docs/latest/matplotlib/geoaxes.html#cartopy.mpl.geoaxes.GeoAxes) object, which is a subclass of `matplotlib.axes` and provides a large number of special cartographics methods.

```python
# Just tell matplotlib to use a specific map projection
ax = plt.axes(projection=ccrs.PlateCarree())
# show the whole earth
ax.set_global()
# ... and plot coastlines
ax.coastlines()

ny_lon, ny_lat = -75, 43
delhi_lon, delhi_lat = 77.23, 28.61

# Objects plotted into GeoAxis can take a "transform" keyword that describes the coordinate system of origin.
# In a geodetic coordinate system, a line between two points lies on a great circle.
plt.plot([ny_lon, delhi_lon], [ny_lat, delhi_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

# A line in the target coordinates obviously plots as a line ...
plt.plot([ny_lon, delhi_lon], [ny_lat, delhi_lat],
         color='red', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(ny_lon - 3, ny_lat - 12, 'New York',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.text(delhi_lon + 3, delhi_lat - 12, 'Delhi',
         horizontalalignment='left',
         transform=ccrs.Geodetic())

plt.show()
```

## Digression: Downloading files using the `requests` module
### Download 2-m temperatures RCM simulation output file from the ENSEMBLES project
[`requests`](http://docs.python-requests.org/en/latest/) is an extremely capablke and convenient http module.
+ setting `stream=True` in `requests.get` together with iterating over `r.iter_content` will download the file
   in chunks. This is important for large files which otherwise could clog the memory. Also "Using iter_content will handle a lot of what you would otherwise have to handle when using Response.raw directly".
+ the [`with`](http://stackoverflow.com/questions/1369526/what-is-the-python-keyword-with-used-for) statement makes file operations simple ans safer. Files are automatically closed, whether the operation succeeds or not.
+ Note that the gzipped file (file extension "gz") is being automatically gunzipped by `requests`.
   

```python
import requests
r = requests.get("http://ensemblesrt3.dmi.dk/data/A1B/ETHZ/MM/ETHZ-CLM_SCN_HadCM3Q0_MM_25km_tas.nc.gz", stream=True)
with open("ETHZ-CLM_SCN_HadCM3Q0_MM_25km_tas.nc", 'wb') as f:
    for chunk in r.iter_content(chunk_size=2**20):
        if chunk:
            f.write(chunk)
            f.flush()
```

## Comfortably dealing with rotated pole grids

Particularly simulation output from Regional Climate Models (RCMS) often comes in a "rotated pole" grid. That is a lat-lon coordinate system which a Pole that has been rotated to a position such that the resulting grid cells in the simulation domain have better numerical properties (are more homogeneous).

Here is a little helper class, derived from the Cartoply ccrs.RotatedPole class, that takes a netCDF file as parameter to construct the proper projection.


```python
class RotGrid(ccrs.RotatedPole):
    """
    A rotated grid from a netCDF-file.

    """
    def __init__(self, ncfile):
        ncdat = nc.Dataset(ncfile, 'r')
        pol_lon = ncdat.variables["rotated_pole"].grid_north_pole_longitude
        pol_lat = ncdat.variables["rotated_pole"].grid_north_pole_latitude
        super(RotGrid, self).__init__(pol_lon, pol_lat)
```

## A custom projection: Swiss Grid
We derive the "old" Swiss coordinate system "CH1903/LV03" as a subclass from the generic Cartopy projection:

```python
class CH1903_LV03(ccrs.Projection):
    """
    The Swiss Grid (CH1903/LV03) projection

    """
    def __init__(self):
        # The next two assignments define the Swiss Gridd - projection
        proj4_params = [('proj', 'somerc'), ('lat_0', '46.95240555555556'),
                        ('lon_0', '7.439583333333333'), ('k_0', 1),
                        ('x_0', 600000), ('y_0', 200000), ('units', 'm')]
        globe = ccrs.Globe(ellipse='bessel', nadgrids='chenyx06etrs.gsb')
        # The next line initializes the parent-class with the projection parameters.
        super(CH1903_LV03, self).__init__(proj4_params, globe=globe)

    @property
    def threshold(self):
        return 1e4

    @property
    def boundary(self):
        x0, x1 = self.x_limits
        y0, y1 = self.y_limits
        return ccrs.sgeom.LineString([(x0, y0), (x0, y1),
                                      (x1, y1), (x1, y0),
                                      (x0, y0)])
    @property
    def x_limits(self):
        return (480000, 850000)

    @property
    def y_limits(self):
        return (62000, 302000)
```

```python
# An instance of a SwissGrid,
# and one of a geodetic (latlon) grid

swiss = CH1903_LV03()
geo = ccrs.Geodetic()
```
