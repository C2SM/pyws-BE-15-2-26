## Preparation
### Import some necessary modules for making nice specialized figures.

The module `lagranto` is part of the Lagrangian Analysis Tool, ["LAGRANTO"](http://www.geosci-model-dev-discuss.net/8/1893/2015/gmdd-8-1893-2015.html) ([SVN](https://svn.iac.ethz.ch/websvn/pub/)).


```python
%pylab inline
rcParams['figure.figsize'] = (10,6)
from scipy import interpolate
from lagranto import Tra
from mpl_toolkits.axes_grid1 import AxesGrid
import matplotlib.gridspec as gridspec
from mpl_toolkits.basemap import Basemap
from matplotlib.collections import LineCollection
```

### Loading trajectories using Tra

```python
filename = 'data/lsl_20070119_12_ana_48'
trajs = Tra(filename)
print trajs
print trajs.shape
```

## Exercise 4.1
### Interpolate PV on evenly spaced pressure level
Interpolate at each 50hPa pressure level between 1000hPa and 200hPa

Hint: Use the scipy.interpolate module

```python
ilevels = np.arange(200,1000,50)
nlev = len(ilevels)
```

```python
def interp(pvalue, intvalue):
    interpfield = interpolate.interp1d(pvalue[::-1], intvalue[::-1], bounds_error=False)
    return interpfield(ilevels)
```

```python
ntrajs = {}
variables = ['Q', 'RH', 'TH', 'PV']
for var in variables:
    ntrajs[var] = np.zeros((trajs.ntra, nlev))
for i, traj in enumerate(trajs):
    for var in variables:
        ntrajs[var][i, :] = interp(traj['P'], traj[var])
```

## Exercise 4.2
### Plot the mean, mean + std, mean - std of PV interpolated on pressure levels
Use the decreasing pressure as x axis
hint:
* use np.nanmean and np.nanstd
* use matplotlib fill_bewteen function

```python
ax = subplot(111)
mean = np.nanmean(ntrajs['PV'], axis=0)
std = np.nanstd(ntrajs['PV'], axis=0)
pstd = mean + std
mstd = mean - std
_ = fill_between(ilevels, pstd, mstd, alpha=0.5, label='+/- standar deviation')
_ = plot(ilevels, np.nanmean(ntrajs['PV'], axis=0), 'r', label='mean')
_ = xticks(ilevels)
ax.invert_xaxis()
legend()
```

## Exercise 4.3
### Plot the same paramaters as above for TH, RH, PV, Q in a 4x4 plot
Hint : use the AxesGrid class from the mpl_toolkits.axes_grid1 module

```python
def plot_mean_std(pos, var):
    ax = subplot(pos)
    mean = np.nanmean(ntrajs[var], axis=0)
    std = np.nanstd(ntrajs[var], axis=0)
    pstd = mean + std
    mstd = mean - std
    mins = np.nanmin(mstd)
    maxs = np.nanmax(pstd)
    ax.set_title(var)
    _ = ax.fill_between(ilevels, pstd, mstd, alpha=0.5, label='+/- standar deviation')
    _ = ax.plot(ilevels, np.nanmean(ntrajs[var], axis=0), 'r', label='mean')
    _ = ax.set_xticks(ilevels)
    ax.invert_xaxis()
    legend()
```

```python
fig = figure(figsize=(20,20))
grid = gridspec.GridSpec(2,2, wspace=0.1, hspace=0.1)
for box, var in zip(grid, variables):
    plot_mean_std(box, var)
```

## Exercise 4.6
### Plot the trajectories on a map
Hint : Use the Basemap class from the mpl_toolkits.basemap module

```python
lllon,urlon = trajs['lon'].min(), trajs['lon'].max()
lllat,urlat = trajs['lat'].min(), trajs['lat'].max()
mapf = Basemap(llcrnrlon=lllon,llcrnrlat=lllat, urcrnrlon=urlon, urcrnrlat=urlat, resolution='i')
```

```python
mapf.drawcoastlines()
mapf.plot(trajs['lon'].T, trajs['lat'].T, latlon=True, color='k')
title('Trajectories')
```

## Exercise 4.7
### Plot the trajectories on a map, coloring them with pressure
Hint : Use the LineCollection class from matplotlib.collections module

```python
mapf.drawcoastlines()
for traj in trajs:
    points = np.array([traj['lon'], traj['lat']]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    lc = LineCollection(segments, cmap=plt.get_cmap('Spectral'))
    lc.set_array(traj['P'])
    plt.gca().add_collection(lc)
colorbar(lc)
```
